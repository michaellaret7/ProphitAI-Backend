```python
@with_session('user')
def visualize_users(session = None):
    users = session.query(User).all()

    objects = []
    for user in users:
        objects.append(serialize_sqlalchemy_obj(user))
    
    return objects

@with_session('user')
def visualize_companies(session = None):
    companies = session.query(Company).all()

    objects = []
    for company in companies:
        objects.append(serialize_sqlalchemy_obj(company))
    
    return objects

@with_session('user')
def visualize_portfolios(session = None):
    portfolios = session.query(Portfolio).all()

    objects = []
    for portfolio in portfolios:
        objects.append(serialize_sqlalchemy_obj(portfolio))
    
    return objects

a = visualize_portfolios()
b = visualize_companies()
c = visualize_users()
```

```python
@with_session('prophit')
def funds(session = None):
    funds = session.query(Fund).all()

    objects = []
    for fund in funds:
        objects.append(serialize_sqlalchemy_obj(fund))
    
    return objects

@with_session('prophit')
def final_positions(session = None):
    final_positions = session.query(FundFinalPosition).all()

    objects = []
    for fund_final_position in final_positions:
        objects.append(serialize_sqlalchemy_obj(fund_final_position))
    
    return objects

@with_session('prophit')
def initial_positions(session = None):
    initial_positions = session.query(FundInitialPosition).all()

    objects = []
    for fund_initial_position in initial_positions:
        objects.append(serialize_sqlalchemy_obj(fund_initial_position))
    
    return objects

a = funds()
b = final_positions()
c = initial_positions()
```
